### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.tail()



```

#1. Análisis de facturación



```python
ingresos = {
    "Tienda 1": tienda1["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

ingresos_df = pd.DataFrame.from_dict(ingresos, orient="index", columns=["Ingresos"])
ingresos_df["Ingresos"] = ingresos_df["Ingresos"].apply(lambda x: f"{x:,.2f}")
ingresos_df
```

```python
import matplotlib.pyplot as plt

```

```python
tiendas = list(ingresos.keys())
ingresos_values = list(ingresos.values())


plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos_values, color='skyblue')
plt.title('Ingresos por Tienda')
plt.ylabel('Ingresos ($)')
plt.xlabel('Tienda')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
prom_t1 = tienda1["Calificación"].mean()
prom_t2 = tienda2["Calificación"].mean()
prom_t3 = tienda3["Calificación"].mean()
prom_t4 = tienda4["Calificación"].mean()

```

```python
print("Promedio tienda 1:", round(prom_t1, 2))
print("Promedio tienda 2:", round(prom_t2, 2))
print("Promedio tienda 3:", round(prom_t3, 2))
print("Promedio tienda 4:", round(prom_t4, 2))
```

```python
promedios = {
    "Tienda 1": prom_t1,
    "Tienda 2": prom_t2,
    "Tienda 3": prom_t3,
    "Tienda 4": prom_t4
}
plt.figure(figsize=(6, 6))
plt.pie(promedios.values(), labels=promedios.keys(), autopct='%1.2f%%', startangle=90)
plt.title("Distribución del promedio de calificaciones por tienda")
plt.axis('equal')  # Mantener el círculo redondo
plt.show()
```
